# 01. Seaborn 소개

`-` 기존 pandas에서 제공하는 그래프보다 다양하고 근사한 그래프를 그릴 수 있다

`-` Statistical Data Visualizaion 이라고 소개하고 있음

`-` 용어 정리

* PDF(확률밀도함수) : 데이터셋의 분포를 나타낸다.

    * 특정 구간에 대한 확률값을 구하고 싶으면 해당 구간의 면적을 구하면 된다.
    
    * 그래프 아래의 모든 면적을 더하면 1이 된다.
    
    * 연속형 변수, 예를 들어 어떤 학생의 키가 173.5일 확률은 0이다. 왜냐하면 면적아래의 넓이를 특정값으로 구할 수 없기 때문이다.
    
    * 주사위, 동전을 던지는 예제에서는 특정 값에 대한 학률을 구할 수 있음!

# 02. KDE plot

`-` 세상에 일어나는 일들은 신기하게도 대부분 비슷한 분포를 띈다..

* 근데 우리가 모은 데이터로 곡선을 그려보면 매끄러운 그래프가 아닌 약간 울긋불긋한 형태가 나온다.

* `KDE(Kernel Density Esitimation) plot`을 이용하면 특정 분포에 가까운 그래프를 그려주기 때문에 부드러운 곡선을 그려준다!

```python
#!pip install seaborn
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/body.csv")
df.head()
```

`-` 기존 그래프와 seaborn 비교

```python
 df["Height"].value_counts().sort_index().plot()
```

`-` 부드럽다... 이게 바로 우리가 배운 확률밀도함수를 나타내는 것임

* 근데 당연히 실제값과 다름, 우리가 가지고 있는 데이터로 추측을 하는 것이기 때문

```python
sns.kdeplot(df["Height"])
```

* 근데 다음과 같은 인자를 설정하면?

```python
sns.kdeplot(df["Height"], bw = 0.05)
```

```python
sns.kdeplot(df["Height"], bw = 0.5)
```

* `bw` 히스토그램으로 치면 몇 개의 구간으로 나누는지 정하는 파라미터이다.

* 너무 과하게 `bw=2` 이런식으로 사용하면 부정확한 값을 출력할 수 있다.

## kde 활용 예시

`-` 기존의 그래프 그리기

```python
df.plot(kind = "hist", y = "Height", bins = 15)
```

`-` `seaborn`을 사용하면 히스토그램위에 `kde`를 얹을 수 있다.

```python
sns.displot(df["Height"], bins = 15, kde = True)
```

`-` boxplot 그리기

*  기존의 방식

```python
df.plot(kind = "box", y = "Height")
```

`-` 오 boxplot에서 kde plot 그리기 $\to$ `sns.violinplot`

* 근데 난 별로...

```python
sns.violinplot(y = df["Height"])
```

`-` 키와 몸무게의 연광성 보기

* 기존의 방식

```python
df.plot(kind = "scatter", x = "Height", y = "Weight")
```

`-` seaborn

```python
sns.kdeplot(df["Height"], df["Weight"])
```

* 등고선처럼 표시가 되어있다?

`-` 각각에 대한 kde그리기

```python
sns.kdeplot(df["Height"])
```

```python
sns.kdeplot(df["Weight"])
```

`-` 아 아래 그래프는 x축에는 키에 대한 kde, y축에는 몸무게에 대한 kde를 그린 것이다.

```python
sns.kdeplot(df["Height"], df["Weight"])
```

`-` 음... 한번에 보자

```python
fig, axes = plt.subplots(1,3, figsize = (12,4))

ax1,ax2,ax3 = axes

sns.kdeplot(df["Height"], ax = ax1)
sns.kdeplot(df["Weight"], ax = ax2)
sns.kdeplot(df["Height"], df["Weight"], ax = ax3)

fig.tight_layout()
```

* 3번쨰 그림 해석 : 선끼리 가까우면 가파르게 올라가는 것, 멀어지면 완만하게 올라가거나 내려가는 것을 의마한다.

* 근데.. 이거 굳이 쓰지말자..해석하는데 머리 아프다..

# 03. LM plot

`-` 키와 몸무게의 연관성을 보기위한 방법

* 어떤 학생의 키가 주어졋을 때 몸무게를 예측하는 그래프를 그려보자

```python
sns.lmplot(data = df, x = "Height", y = "Weight")
```

* 근데 거의 대부분 점들이 선이랑 떨어져 있다. 즉 저 직선이 뭔가 예측을 잘하는 것 같지 않음

* 애초에 우리가 수집한 데이터에서 키와 몸무게간 그렇게 연관성이 있어 보이지 않음

# 04. 카테고리별 시각화

```python
df = pd.read_csv("data/laptops.csv")
df.head()
```

`-` category plot의 약자!

* os별 가격을 확인

```python
sns.catplot(data = df, x = "os", y = "price", kind = "box")
```

```python
sns.catplot(data = df, x = "os", y = "price", kind = "violin")
```

`-` 근데 각 범주별 데이터의 개수를 모르니까 그렇게 신뢰할 수 있는 그래프인지 모르겠음...

* 다음과 같이 `strip`을 이용하자

```python
sns.catplot(data = df, x = "os", y = "price", kind = "strip")
```

`-` 근데 또 궁금한거 $\to$ 컴퓨터 프로세스를 만드는 기업별로 가격을 확인하고 싶음

```python
sns.catplot(data = df, x = "os", y = "price", kind = "strip", hue = "processor_brand")
```

* 오 `intel`을 쓰는 노트북이 더 많은 것 같다..

* 근데 문제점 : 점들이 많이 모여있는 부분은 많이 겹쳐있어서 보기 좀 불편

```python
sns.catplot(data = df, x = "os", y = "price", kind = "swarm", hue = "processor_brand")
```

`-` 흠.... 뭔가 되게 징그럽다... 이거를 `report`로 넣기엔 개인차가 많이 갈릴 것 같음
