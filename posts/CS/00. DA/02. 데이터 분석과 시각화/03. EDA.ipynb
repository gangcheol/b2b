# 상관관계 분석

```python
df = pd.read_csv("data/young_survey.csv")
df.head()
```

```python
music = df.iloc[: , : 19]
```

```python
sns.heatmap(music.corr())
```

`-` 나이와 연관성이 높은 변수 내림차순 정렬

# 클러스터 분석

`-` 클러스터 : 어떤 뭉쳐있는 무리

* 즉, 다양한 데이터를 가지고, 비슷한 성향을 가진 개체끼리 분석하는 기법

```python
interests = df.loc[: , "History": "Pets"]
interests.head()
```

```python
corr = interests.corr()
```

`-` 역사를 좋아하는 사람을 기준으로 연관성이 높은 변수 산출

```python
corr["History"].sort_values(ascending = False)
```

`-` 근데, 한 컬럼 하나하나 일일이 확인하는 것은 비효율적임

```python
sns.clustermap(corr, figsize = (12,7))
```

### Q1

* (X) : 20대와 30대가 가장 많다.

### Q2

`-` 가장 높은 요금을 낸 사람은 30대이다.

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "plotly_mimetype+notebook_connected"
```

```python
df["Age"].describe()
```

```python
import numpy as np
```

```python
bins = [-np.inf, 19,29,39,49,59, np.inf]
labels = ["20대 미만", "20대", "30대", "40대", "50대", "60대 이상"] 
```

```python
df["Age_label"] =  pd.cut(df["Age"], bins = bins, labels = labels)
```

```python
import warnings

warnings.filterwarnings(action = "ignore")
```

```python
df.groupby("Age_label")["Fare"].sum()
```

`-` (O) : 30대가 가장 많은 요금을 지불했음

### Q6

`-` (X) : 나이가 어릴수록 생존율이 높다.

`-` 저렇게 보면 나이가 어릴수록 생존율이 높다고 볼 수 있음

* 그러나 분포의 그렇게 큰 차이가 보이지 않으므로 정확한 답을 내리기 어렵다....

```python
sns.stripplot(data=df, x="Survived", y="Age", hue = "Survived")
```

### Q7

`-` (O) :  나이보다 성별이 생존율에 더 많은 영향을 미친다.

```python
df["Sex"] = df["Sex"].replace(["male","female"],[0,1])
```

```python
df.select_dtypes("number").corr()["Survived"].sort_values(ascending = False)
```

```python
# 생존 여부에 따른 나이 및 성별 분포
sns.stripplot(data=df, x="Survived", y="Age", hue="Sex")
```

`-` 나이 분포는 비슷한 데 비해, 성별의 분포는 확연히 차이가 난다.

* 따라서, 생존율은 나이보다는 확실히 성별에 영향을 많이 받았다는 걸 알 수 있습니다.
